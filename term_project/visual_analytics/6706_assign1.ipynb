```
import numpy as np
import matplotlib.pyplot as ax
 
fig, plt = ax.subplots(figsize=(10,6))
# set width of bar
barWidth = 0.25
 
# set height of bar
FullMesh = [32.314, 22.865, 20.281]
DropAlgorithm = [64.585, 86.985, 52.004]
 
# Set position of bar on X axis
r1 = np.arange(3)
r2 = r1 + barWidth
 
# Make the plot
plt.bar(r1, FullMesh, color='#1f6d5f', width=barWidth, edgecolor='white', label='Full Mesh')
plt.bar(r2, DropAlgorithm, color='#957f2d', width=barWidth, edgecolor='white', label='Drop Algorithma')
 
# Add xticks on the middle of the group bars
ax.xlabel('OpenFlow Links', fontweight='bold')
ax.ylabel('Round Trip Time (ms : Delay)', fontweight='bold')
ax.xticks([r + barWidth for r in range(len(FullMesh))], ['A-D', 'A-B', 'B-C'])
 
# Create legend & Show graphic
plt.legend()
fig.suptitle('Network RTT Avg Comparsion Between Full Mesh and Drop Algorithm', fontsize=16)
ax.show()
```
